# **Copy/Save/Load**

```python
a = np.array([1, 2, 3])
b = a.copy()
b[0] = 100
print(a, b)
```

```python
np.save('my_array', a)
loaded = np.load('my_array.npy')
print(loaded)
```

# **Matrix Operation**

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
print(np.matmul(a, b))
```

```python
from numpy.linalg import inv
a = np.array([[1, 2], [3, 4]])
print(inv(a))
```

```python
from numpy.linalg import det
print(det(a))
```

```python
print(np.trace(a))
```
